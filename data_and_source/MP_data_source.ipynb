```python
import pandas as pd
import numpy as np
from pymatgen import MPRester
```

# Use pymatgen MPRester api quickly get the properties in the request set

```python
a = MPRester("thsu@gmial.com/sutianhao@shu.edu.cn") # your own materials project API

results = a.query(criteria={},properties=["magnetic_type","task_id"])
#results = a.query(criteria={"task_id": "mp-999204"}, properties=["magnetic_type","pretty_formula", "full_formula", "unit_cell_formula", "input.potcar_spec.titel","task_id"])
#print(results)
```

```python
results
```

```python
df = pd.DataFrame(results)
df
```

```python
df.to_csv("./magnetic_type.csv")
```

# You can also get the specified properties in this way

```python
results_mag = a.query(criteria={"magnetic_type": "FM"}, properties=["task_id"])
task_id=pd.DataFrame(results_mag, columns=["task_id"])
task_id.to_csv("FM.csv")
```
